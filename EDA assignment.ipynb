**Frequency Table**

```python
unique_labels=df['Gender'].unique()
count=[]
for i in unique_labels:
    con=df['Gender']==i
    count.append(len(df[con]))
    Gender_df=pd.DataFrame(zip(unique_labels,count),
                          columns=['Gender','count'])
Gender_df
```

```python
Gender_df.to_csv('Gender_df.csv',index=False)
```

```python
Gender_vc=df['Gender'].value_counts()
l1=Gender_vc.keys()
l2=Gender_vc.values
Gender_vc_df=pd.DataFrame(zip(l1,l2),
                         columns=['Gender','count'])
Gender_vc_df
```

```python
Gender_vc_df
```

```python
plt.figure(figsize=(4,2))
plt.bar('Gender',
       'count',data=Gender_vc_df)
plt.xlabel("candidate")
plt.ylabel("count")
plt.title("Bar chart")
plt.savefig("Gender_bar.jpg")
plt.show
```

```python
plt.figure(figsize=(4,2))
l=df['Gender'].value_counts().keys()
sns.countplot(data=df,x='Gender',order=l)
plt.xlabel('Candidate')
plt.ylabel('Count')
plt.title('Bar Chart')
plt.savefig('Gender_bar.jpg')
plt.show()
```
